```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
```

```python
lambda_L = 4
lambda_H = 7.5
mu_L = 0.2
mu_H = 0.286
N = 40
n = 0 
new_time = 0 
n_A = 0
n_D = 0
arrival_time_list = []
departure_time_list = []
n_list = []
time_list = []
```

```python
x_HL = lambda_H / mu_L
x_LH = lambda_L / mu_H
x_HH = lambda_H / mu_H
x_LL = lambda_L / mu_L
```

```python
na_list = np.arange(int(x_HL) + 1, N + 1)
ns_list = np.arange(int(x_HL) + 1, N + 1)
```

```python
def remove_values_from_list(the_list, val):
    return [value for value in the_list if value != val]
```

#First scenario: $c_a=1, c_s=2, c_w=3$  

#Greedy algorithm

```python
ca = 1
cs = 2
cw = 3
```

```python
servers = []
for _ in range(N):
    servers.append(0)
```

```python
if max(ca,cs,cw) == ca:
    
    na = np.max(na_list)
    
    if max(cs,cw) == cs:
        
        
        ns = np.max(ns_list)
        
        
    else:
        
        ns = np.min(ns_list)
    
elif max(ca,cs,cw) == cs:

    ns = np.max(ns_list)
    
    
    if max(ca,cw) == ca:
        
        na = np.max(na_list)
        
        
    else:
        
        na = np.min(na_list)
        
        
        
elif max(ca,cs,cw) == cw:
    
    
    if max(ca,cs) == cs:
        
        
        na = np.min(na_list)
        ns = np.max(ns_list)
        
    else:
    
    
        na = np.max(na_list)
        ns = np.min(ns_list)
    
    
print("na : " ,na)
print("ns : " ,ns)
```

```python
for _ in range (10000): 
    
    
    if n < na and n <ns:
        
        lambda_ = lambda_H
        mu_ = mu_L
        
        
    elif n > na and n > ns:
        
        lambda_ = lambda_L
        mu_ = mu_H
        
        
    elif n < ns and n > na :
        
        lambda_ = lambda_L
        mu_ = mu_L
        
    elif n > ns and n < na:
        
        lambda_ = lambda_H
        mu_ = mu_H
        
    
    
    
    time = np.random.exponential(1/((n*mu_)+ lambda_))
    #print(time)
    new_time += time
    time_list.append(new_time)
    #print(new_time)

    probability_of_arrival = lambda_ / ((n*mu_)+ lambda_)
    probability_of_departure = (n*mu_) / ((n*mu_)+ lambda_)
    events = ["arrival","departure"]
    distributions = [probability_of_arrival,probability_of_departure]
    random_events = np.random.choice(events, 1, True, distributions)
    #print(random_events)

    if random_events == "arrival":
        n += 1
        n_A += 1
        arrival_time = new_time
        arrival_time_list.append(arrival_time)
        servers.sort()
        servers[0] += 1

    elif random_events == "departure":

        n -= 1
        n_D += 1
        departure_time = new_time
        departure_time_list.append(departure_time)
        number_of_empty_servers = servers.count(0)
        alternative_servers = remove_values_from_list(servers,0)
        random_index = random.randint(0,len(alternative_servers)-1)
        alternative_servers[random_index] -= 1
        for _ in range(number_of_empty_servers):
            alternative_servers.append(0)
        servers = alternative_servers
        
    #print(n)
    n_list.append(n)
    
        
    new_time = time
```

```python
# E (X - N) +
element_list = []
for element in n_list:
    ele  = max(0,element - N)
    element_list.append(ele)
expected_value = statistics.mean(element_list)
main_len = len(element_list)
```

```python
element_list.sort()
num = element_list.count(0)
for _ in range(num):
    element_list.remove(element_list[0])
    
secondary_len = len(element_list)
```

```python
probability_of_wait = secondary_len / main_len
```

```python
probability_of_wait
```

```python
larger_elements = [element for element in n_list if element > na]
number_of_elements = len(larger_elements)
probability_of_admission_control = number_of_elements / len(n_list)
print(probability_of_admission_control)
```

```python
larger_elements = [element for element in n_list if element > ns]
number_of_elements = len(larger_elements)
probability_of_speedup = number_of_elements / len(n_list)
print(probability_of_speedup)
```

```python
main_cost = (ca*probability_of_admission_control) + (cs* probability_of_speedup) + (cw*expected_value )
```

```python
main_cost
```

#Optimal policy

```python
main_n_list = []
probability_of_admission_control_list = []
probability_of__speedup_list = []
```

```python
servers = []
for _ in range(N):
    servers.append(0)
```

```python
for i in na_list:
    
    for j in ns_list:
        
        
        N = 40
        n = 0 
        new_time = 0 
        n_A = 0
        n_D = 0
        arrival_time_list = []
        departure_time_list = []
        n_list = []
        time_list = []

        number_of_elements_for_admission_control = 0
        number_of_elements_for_speedup = 0


        for _ in range (10000): 
    
    
            if n < i and n <j:

                lambda_ = lambda_H
                mu_ = mu_L


            elif n > i and n > j:

                lambda_ = lambda_L
                mu_ = mu_H


            elif n < j and n > i :

                lambda_ = lambda_L
                mu_ = mu_L

            elif n > j and n < i:

                lambda_ = lambda_H
                mu_ = mu_H




            time = np.random.exponential(1/((n*mu_)+ lambda_))
            #print(time)
            new_time += time
            time_list.append(new_time)
            #print(new_time)

            probability_of_arrival = lambda_ / ((n*mu_)+ lambda_)
            probability_of_departure = (n*mu_) / ((n*mu_)+ lambda_)
            events = ["arrival","departure"]
            distributions = [probability_of_arrival,probability_of_departure]
            random_events = np.random.choice(events, 1, True, distributions)
            #print(random_events)

            if random_events == "arrival":
                n += 1
                n_A += 1
                arrival_time = new_time
                arrival_time_list.append(arrival_time)
                servers.sort()
                servers[0] += 1

            elif random_events == "departure":

                n -= 1
                n_D += 1
                departure_time = new_time
                departure_time_list.append(departure_time)
                number_of_empty_servers = servers.count(0)
                alternative_servers = remove_values_from_list(servers,0)
                random_index = random.randint(0,len(alternative_servers)-1)
                alternative_servers[random_index] -= 1
                for _ in range(number_of_empty_servers):
                    alternative_servers.append(0)
                servers = alternative_servers

            #print(n)
            n_list.append(n)
            
            
            new_time = time
            
            
            if n > i :
                number_of_elements_for_admission_control +=1

            
            if n > j :
                number_of_elements_for_speedup += 1 
            
        probability_of_admission_control = number_of_elements_for_admission_control / len(n_list)
                
        probability_of_speedup_list = float(number_of_elements_for_speedup / len(n_list))
            
            
        probability_of_admission_control_list.append(probability_of_admission_control)

        probability_of__speedup_list.append(probability_of_speedup_list)

        
        main_n_list.append(n_list)
```

```python
main_n_list
```

```python
# E (X - N) +
main_element_list = []
for i in range (len(main_n_list)):
    for element in main_n_list[i]:
        element_list = []
        element = max(0,element - N)
        element_list.append(element)
    expected_value = statistics.mean(element_list)
    main_element_list.append(expected_value)
```

```python
main_cost_list = []

for i in range(len(main_element_list)):
    main_cost = cw * main_element_list [i] + ca * probability_of_admission_control_list[i] + cs * probability_of__speedup_list[i]
    main_cost_list.append(main_cost)
    min_cost = np.min(main_cost_list)
    index_min = np.argmin(main_cost_list)
    

print(min_cost)
print(index_min)
```

```python
len_n = len(main_n_list)
len_na = len(na_list)
len_ns = len(ns_list)
number_of_categories = len_n / len_ns
number_of_elements = len_n / len_na
if index_min == 0:
    na = na_list[0]
    ns = ns_list[0]
    
    
else:
    
    categories = np.ceil(index_min / len_ns)
    na = na_list[int(categories) - 1]
    element = index_min - ( int(categories) - 1 )* ( len_ns )
    ns = ns_list[element]

print("na :", na)
print("ns : ", ns)
```
